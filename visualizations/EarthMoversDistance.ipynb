```python
from one.api import ONE
import numpy as np

# Instantiate the ONE API
one = ONE(base_url='https://openalyx.internationalbrainlab.org', password='international')
```

```python
spks = one.load_dataset('19e66dc9-bf9f-430b-9d6a-acfa85de6fb7', 'spikes.times.npy', collection='alf/probe00/pykilosort')
clusters = one.load_dataset('19e66dc9-bf9f-430b-9d6a-acfa85de6fb7', 'spikes.clusters.npy', collection='alf/probe00/pykilosort')
```

```python
import matplotlib.pyplot as plt
for i in range(0, 795):
    try:
        plt.hist(np.log(np.diff(spks[clusters==i])),bins=100,alpha=0.05,)
    except:
        pass
plt.show()
```

```python
import matplotlib.pyplot as plt
for i in range(0, 795):
    try:
        plt.hist(np.log(np.diff(spks[clusters==i])),bins=100,alpha=0.08,range=(-7.5,1))
    except:
        pass
plt.show()
```

```python
import matplotlib.pyplot as plt
hist_lst=[]
for i in range(0, 795):
    try:
        a=np.histogram(np.log(np.diff(spks[clusters==i])), bins=100,range=(-7.5,1))
        hist_lst.append(a[0])
    except:
        pass
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=10).fit_transform(np.array(hist_lst))
```
