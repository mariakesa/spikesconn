```python
clusters=b=one.load_dataset('19e66dc9-bf9f-430b-9d6a-acfa85de6fb7', 'clusters', collection='alf/probe00/pykilosort')
```

```python
# specify experiment ID
eid = '19e66dc9-bf9f-430b-9d6a-acfa85de6fb7'

# load cluster IDs and corresponding channels
clusters = one.load_object(eid, 'clusters')
print(clusters)
waveform_channels = clusters['waveformsChannels']

# load waveforms for each cluster
waveforms = one.load_dataset(eid, 'clusters.waveforms')

# plot EAP waveform for each cluster
for i, waveform_channel in enumerate(waveform_channels):
    channel_indices = np.squeeze(waveform_channel)
    cluster_waveforms = waveforms[i, :, channel_indices]
    mean_waveform = np.mean(cluster_waveforms, axis=0)
    timepoints = np.arange(-mean_waveform.shape[0]/2, mean_waveform.shape[0]/2)
    plt.plot(timepoints, mean_waveform)
    plt.xlabel('Time (ms)')
    plt.ylabel('Voltage (uV)')
    plt.title('EAP waveform for cluster {}'.format(i+1))
    plt.show()
```
